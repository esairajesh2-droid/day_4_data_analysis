```python
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.DataFrame({
    'Language':['Python','Java','JavaScript','C++','C'],
    'Users':[45,55,15,10,5],
    'colors':['red','blue','green','orange','yellow']
})
x=df.Language
y=df.Users
plt.pie(y,labels=x,autopct='%1.0f%%',startangle=90,colors=df.colors,radius=1.5,wedgeprops={'edgecolor':'white','linewidth':2},textprops={'fontsize':15},pctdistance=0.8)
plt.legend()
plt.show()
```

```python
#showing there is positive correlation
df1=pd.DataFrame({
    'height':[160,165,170,180,185,190],
    'weight':[55,60,65,72,78,84]
})
x=df1.height
y=df1.weight
plt.figure(figsize=(8,5))
plt.title("Height and weight")
plt.scatter(x,y,marker='o',edgecolors='red',linewidths=1,color='orange',s=120)
plt.xlabel('height',fontsize=25,fontstyle='italic',color='red')
plt.ylabel('weight',fontsize=25,fontstyle='italic',color='red')
plt.gcf().set_facecolor(color='yellow')
plt.gca().set_facecolor(color='skyblue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#showing there is negative correlation
df2= pd.DataFrame({
    "Age": [18, 22, 26, 30, 34, 38, 42],
    "Speed": [28, 27, 26, 24, 22, 20, 18]
})
x=df2.Age
y=df2.Speed
plt.figure(figsize=(8,5))
plt.title("Age and Speed")
plt.scatter(x,y,marker='o',edgecolors='blue',linewidths=1,s=120)
plt.xlabel('Age',fontsize=25,fontstyle='italic',color='blue')
plt.ylabel('Speed',fontsize=25,fontstyle='italic',color='blue')
plt.gcf().set_facecolor(color='pink')
plt.gca().set_facecolor(color='skyblue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#Showing there is no correlatiom
df3 = pd.DataFrame({
    "o": [10, 20, 30, 40, 50, 60, 70],
    "p": [35, 10, 55, 25, 60, 20, 45]
})
x=df3.o
y=df3.p
plt.figure(figsize=(8,5))
plt.title("X and Y")
plt.scatter(x,y,marker='o',edgecolors='red',linewidths=1,s=120)
plt.xlabel('x',fontsize=14,fontstyle='italic',color='red')
plt.ylabel('Y',fontsize=14,fontstyle='italic',color='red')
plt.gcf().set_facecolor(color='red')
plt.gca().set_facecolor(color='skyblue')
plt.grid(True)
plt.tight_layout()
plt.show()
```
