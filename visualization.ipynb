```python
#importing the dataset and required libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\esair\OneDrive\Desktop\thelast\fifa_world_cup_2026_player_performance.csv")
```

```python
df_player=df[df['player_name']=='Lionel Martinez']
```

```python
player=df_player.sort_values(by='match_date',ascending=True).head(10)
print(player)
```

```python
#Dashboard for Lionel Martinez's performance in the last 10 matches
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.gcf().set_facecolor('#0B0F1A')

# Graph 1
plt.subplot(1,2,1)
#total minutes palyed against each opponent
x=player.opponent_team
y=player.minutes_played
plt.title("total minutes played by Lionel Martinez",color='#ffd54f',fontsize=14)
plt.plot(x,y,marker='o',markerfacecolor='#FFD54F',markeredgecolor='#000000',linestyle='solid',color='orange')
plt.xlabel('Opponent Team',color='#FFFFFF',fontsize=12,fontstyle='italic')
plt.ylabel('Minutes played',color='#FFFFFF',fontsize=12,fontstyle='italic')
plt.xticks(color='#ffd54f',fontsize=10,rotation=90)
plt.yticks(color='#ffd54f',fontsize=10)
plt.tick_params(axis='x',pad=8)
plt.tick_params(axis='y',pad=8)
plt.gca().set_facecolor(color='#0b0b0b')
plt.grid(color='#444444',axis='y')

# Graph 2
plt.subplot(1,2,2)

#successfull passes against opponent teams
x=player.opponent_team
y=player.successful_passes
plt.title('Successful passes Against Opponent Teams',color='#FF00FF',fontsize=14,fontstyle='italic')
plt.bar(x,y,color='#00F5FF',width=0.8,edgecolor='#39FF14')
plt.xlabel('Opponent Team',fontsize=12,color='#FF00FF',fontstyle='italic',labelpad=15)
plt.ylabel('Successfull passes',color='#FF00FF',fontsize=12,fontstyle='italic',labelpad=15)
plt.xticks(rotation=90,color='#00FFFF',fontsize=10)
plt.yticks(color='#00FFFF',fontsize=10)
plt.tick_params(axis='x',pad=15)
plt.tick_params(axis='y',pad=15)
plt.gca().set_facecolor(color='#1A1A2E')
plt.grid(axis='y',color='#5a1846',linewidth=2)
plt.suptitle(
    "Lionel Martinez Performance Dashboard",
    fontsize=18,
    color='white'
)


plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
